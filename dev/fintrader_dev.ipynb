```python
# Import libraries.
import pandas as pd
from pathlib import Path
```

```python
# CSV paths
amzn_csv = Path(
        "./resources/amazon.csv"
        )
sox_csv = Path(
        "./resources/SOX_2015_2020.csv"
        )

# Create list of paths.
csv_dict = {
    "amzn": amzn_csv,
    "sox": sox_csv
}

# Review list.
csv_dict["sox"]
```

```python
pd.read_csv(csv_dict["amzn"]).head()
```

```python
# Review dict.
for k, v in csv_dict.items():
    print(
        f"{k} |==> {v}"
    )
```

```python
# Read csv into dataframe by looping.
df = {}
for k, v in csv_dict.items():

    # Pandas dataframe.
    csv_df = pd.read_csv(csv_dict[k])
    df[k] = csv_df
```
